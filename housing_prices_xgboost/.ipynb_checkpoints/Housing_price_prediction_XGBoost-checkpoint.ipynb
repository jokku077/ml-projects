```python
#importing required packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn.datasets
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor


from sklearn.metrics import accuracy_score
```

```python
h_data = pd.read_csv("HousingData.csv")
```

```python
h_data.head()
```

```python
h_data.shape
```

```python
h_data.isnull().sum() #null values in the dataset
```

we have found out that there are missing values


we fill missing values with mean of the attribute. 

```python
#replacing null values with attribute mean
h_data['CRIM'].fillna(h_data['CRIM'].mean(), inplace=True)
h_data['ZN'].fillna(h_data['ZN'].mean(), inplace=True)
h_data['INDUS'].fillna(h_data['INDUS'].mean(), inplace=True)
h_data['CHAS'].fillna(h_data['CHAS'].mean(), inplace=True)
h_data['AGE'].fillna(h_data['AGE'].mean(), inplace=True)
h_data['LSTAT'].fillna(h_data['LSTAT'].mean(), inplace=True)

#note : we are only doing this for attributes without null values
```

```python
h_data.isnull().sum()
```

we have taken care of all null value

```python
h_data.describe()
```

Understanding correlation between data

```python
correlation = h_data.corr()
```

```python
# constructing heat map for correlation
plt.figure(figsize=(12,12))
sns.heatmap(correlation, cbar = True, fmt = '.1f', annot = True, cmap = 'Greens', )
  # cbar: whether we need color bar,  fmt: how many values after decimal point
  # annot = True means that numbers should be dispalyed within each cell
  # cmap specifies our color scheme for the map
```

<h3>Splitting data into labels and input</h3>

```python
X = h_data.drop(columns = 'MEDV', axis = 1)
Y = h_data['MEDV']
```

<h3>splitting into training and test set</h3> 

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.2, random_state=2)
```

<h3>training the model</h3>

```python
model = XGBRegressor()

model.fit(X_train,Y_train)
```

<h3>checking model accuracy</h3>

```python
model_predictions = model.predict(X_test)
#print(model_predictions)
score = accuracy_score(model_predictions, Y_test)
print(score)
```
